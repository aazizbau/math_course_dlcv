## 1. Non-Convexity in High Dimension
In deep learning, saddles dominate and wide basins are easier to reach than sharp isolated minima.


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
sharp = x**4 + 0.1*x**2
flat = 0.2*x**4 + 0.1*x**2

plt.plot(x, sharp, label="Sharp Basin")
plt.plot(x, flat, label="Flat Basin")
plt.legend()
plt.title("Sharp vs Flat Minima")
plt.show()
````

## 1. Double Descent (Conceptual)
Test error can decrease, rise near interpolation, then decrease again as capacity grows.


````python
import numpy as np
import matplotlib.pyplot as plt

capacity = np.linspace(1, 100, 200)
error = 1/(capacity) + 0.02*(capacity-50)**2/1000

plt.plot(capacity, error)
plt.axvline(50, linestyle="--", label="Interpolation Threshold")
plt.title("Conceptual Double Descent")
plt.xlabel("Model Capacity")
plt.ylabel("Test Error")
plt.legend()
plt.show()
````

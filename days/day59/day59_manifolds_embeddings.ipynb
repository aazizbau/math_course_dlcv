## 1. Manifold Intuition
Data lies on a low-dimensional surface embedded in high-dimensional space.


````python
import numpy as np
import matplotlib.pyplot as plt

# Simple swiss-roll-like data
rng = np.random.default_rng(0)
t = 1.5 * np.pi * (1 + 2 * rng.random(500))
x = t * np.cos(t)
y = 21 * rng.random(500)
z = t * np.sin(t)

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z, s=3)
ax.set_title('Swiss Roll Manifold')
plt.show()
````

## 1. Invariance and Augmentation
Augmentation expands the data distribution with label-preserving transforms.


````python
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((100, 100))
img[30:70, 40:60] = 1

plt.figure(figsize=(6,2))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(1,3,2)
plt.imshow(np.roll(img, 10, axis=1), cmap='gray')
plt.title('Translated')

plt.subplot(1,3,3)
plt.imshow(np.rot90(img), cmap='gray')
plt.title('Rotated')
plt.show()
````

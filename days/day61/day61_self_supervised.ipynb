## 1. Contrastive SSL
Positive pairs are augmented views of the same sample; negatives are other samples.


````python
import numpy as np

z1 = np.array([0.8, 0.1])
z2 = np.array([0.75, 0.15])
sim = np.dot(z1, z2) / (np.linalg.norm(z1) * np.linalg.norm(z2))
print('Similarity:', sim)
````

## 1. Train vs Test Loss
Optimization minimizes training loss, while generalization depends on behavior under unseen samples.


````python
train_loss = 0.01
test_loss = 0.14
print("Generalization gap:", test_loss - train_loss)
````

## 2. Capacity and Overfitting
High-capacity models can fit training data perfectly, but that alone does not ensure robust test behavior.


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 20)
y = x**2 + np.random.randn(20) * 0.1
coeffs = np.polyfit(x, y, 15)
poly = np.poly1d(coeffs)
x_test = np.linspace(-1, 1, 200)
plt.scatter(x, y)
plt.plot(x_test, poly(x_test), label="High Capacity Fit")
plt.legend()
plt.title("Overfitting Example")
plt.show()
````

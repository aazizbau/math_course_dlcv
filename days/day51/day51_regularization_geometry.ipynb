## 1. L2 vs L1 Geometry
L2 imposes a spherical constraint, L1 imposes a diamond-shaped constraint.


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, x)
L1 = np.abs(X) + np.abs(Y)
L2 = X**2 + Y**2

plt.contour(X, Y, L1, levels=[1], colors='r')
plt.contour(X, Y, L2, levels=[1], colors='b')
plt.title('L1 (Red) vs L2 (Blue)')
plt.axis('equal')
plt.show()
````

## 2. Weight Decay Update
Weight decay adds a pull toward zero to each gradient step.


````python
theta = 2.0
lr = 0.1
decay = 0.2
vals = []

for _ in range(30):
    theta = theta - lr * decay * theta
    vals.append(theta)

plt.plot(vals)
plt.title('Weight Decay')
plt.show()
````

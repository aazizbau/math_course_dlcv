## 1. Shift Types
- Covariate shift: `p(x)` changes
- Label shift: `p(y)` changes
- Concept shift: `p(y|x)` changes


````python
import numpy as np
import matplotlib.pyplot as plt

train = np.random.normal(0, 1, 1000)
test = np.random.normal(2, 1, 1000)

plt.hist(train, bins=30, alpha=0.5, label="Train")
plt.hist(test, bins=30, alpha=0.5, label="Test")
plt.legend()
plt.title("Covariate Shift Example")
plt.show()
````

## 1. Aleatoric vs Epistemic
Aleatoric = irreducible noise; Epistemic = lack of knowledge.


````python
import numpy as np

preds = np.array([
    [0.7, 0.2, 0.1],
    [0.6, 0.25, 0.15],
    [0.8, 0.15, 0.05],
])
mean_pred = preds.mean(axis=0)
epistemic = preds.var(axis=0)
print('Mean:', mean_pred)
print('Epistemic variance:', epistemic)
````

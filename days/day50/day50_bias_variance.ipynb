## 1. Bias vs Variance
High bias means systematic underfitting; high variance means sensitivity to small data changes.

\[ E[(y-\hat f(x))^2] = Bias^2 + Variance + Noise \]


## 2. Model Capacity
Capacity reduces bias but increases variance. Regularization and data augmentation reduce variance.


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
true = x**3
high_bias = 0.5 * x
high_variance = true + 0.2 * np.random.randn(len(x))

plt.plot(x, true, label='True function', linewidth=2)
plt.plot(x, high_bias, label='High bias')
plt.plot(x, high_variance, label='High variance', alpha=0.7)
plt.legend()
plt.title('Bias vs Variance')
plt.show()
````

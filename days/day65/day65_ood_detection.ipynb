## 1. OOD Core Idea
OOD means inputs are outside the support of training data; confidence alone is unreliable.


````python
import numpy as np
import matplotlib.pyplot as plt

in_data = np.random.normal(0, 1, (200, 2))
ood_data = np.random.normal(5, 1, (50, 2))
plt.scatter(in_data[:,0], in_data[:,1], alpha=0.5, label="In-Distribution")
plt.scatter(ood_data[:,0], ood_data[:,1], alpha=0.8, label="OOD")
plt.legend()
plt.title("In vs Out of Distribution")
plt.show()
````
